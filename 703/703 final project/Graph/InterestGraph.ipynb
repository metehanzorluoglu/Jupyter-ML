```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro

df = pd.read_csv('CleanElementry.csv', encoding = 'latin-1')

Pre = 0
Post = 0
Eq = 0

for index, row in df.iterrows():
    if row['Pre_RI_Interest'] > row['Post_RI_Interest']:
        Pre += 1
    elif row['Pre_RI_Interest'] == row['Post_RI_Interest']:
        Eq += 1
    else:
        Post += 1

print(f"Equal count: {Eq}")
print(f"Pre count: {Pre}")
print(f"Post count: {Post}")
```

```python
labels = [f'Pre > Post ({Pre})', f'Post > Pre ({Post})', f'Post = Pre ({Eq})']
counts = [Pre, Post, Eq]

plt.bar(labels, counts)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Comparison of Pre and Post counts')
plt.show()
```

```python
df['Post_RI_Interest'].plot(kind='hist', bins=5, edgecolor='black')  # Customize bins, color, etc.
plt.title('Histogram of Post_RI_Interest')
plt.xlabel('Post_RI_Interest')
plt.ylabel('Frequency')
plt.show()
```

```python
# Shapiro-Wilk Test
stat, p_value = shapiro(df['Post_RI_Interest'])

# Interpretation
def is_normal(p_value):
    alpha = 0.05
    if p_value > alpha:
        print("Sample looks normal (fail to reject H0)")
    else:
        print("Sample does not look normal (reject H0)")

print("Shapiro-Wilk Test Statistic for Post_RI_Interest:", stat)
print("p-value:", p_value)
```

```python
is_normal(p_value)
```

```python
# Shapiro-Wilk Test
stat, p_value = shapiro(df['Pre_RI_Interest'])

print("Shapiro-Wilk Test Statistic for Pre_RI_Interest:", stat)
print("p-value:", p_value)
```

```python
# Interpretation
is_normal(p_value)
```
