```python
import numpy as np
import matplotlib.pyplot as plt

# Data points (X: age in months, Y: height in inches)
X_data = np.array([0, 1, 2, 3, 4, 5, 6, 7])
Y_data = np.array([19.75, 21.4, 23.5, 23.1, 23.8, 25, 26.4, 27.25])

# Create the design matrix X
X_design = np.vstack([np.ones_like(X_data), X_data]).T

X_design
```

```python
# Calculate w using the normal equation w = (X^T X)^(-1) X^T Y
w = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y_data

w
```

```python
# Scatter plot of the dataset
plt.scatter(X_data, Y_data, color='blue', label='Data Points')

# Plot the fitted line
X_fit = np.linspace(0, 7, 100)
Y_fit = w[0] + w[1] * X_fit
plt.plot(X_fit, Y_fit, color='red', label=f'Fitted Line: Y = {w[0]:.2f} + {w[1]:.2f}X')

# Labels and title
plt.xlabel('Age (months)')
plt.ylabel('Height (inc)')
plt.title('Age vs Height with Fitted Line')
plt.legend()
plt.grid(True)

plt.show()
```
