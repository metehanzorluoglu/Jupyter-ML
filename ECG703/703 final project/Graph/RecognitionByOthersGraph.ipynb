```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro

df = pd.read_csv('CleanElementry.csv', encoding = 'latin-1')

Pre = 0
Post = 0
Eq = 0

preColumn = 'Pre_RI_Recognition_ByOthers'
postColumn = 'Post_RI_Recognition_ByOthers'

for index, row in df.iterrows():
    if row[preColumn] > row[postColumn]:
        Pre += 1
    elif row[preColumn] == row[postColumn]:
        Eq += 1
    else:
        Post += 1

print(f"Equal count: {Eq}")
print(f"Pre count: {Pre}")
print(f"Post count: {Post}")
```

```python
labels = [f'Pre > Post ({Pre})', f'Post > Pre ({Post})', f'Post = Pre ({Eq})']
counts = [Pre, Post, Eq]

plt.bar(labels, counts)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Comparison of Pre and Post counts')
plt.show()
```

```python
df[preColumn].plot(kind='hist', bins=5, edgecolor='black')  # Customize bins, color, etc.
plt.title('Histogram of ' + preColumn)
plt.xlabel(preColumn)
plt.ylabel('Frequency')
plt.show()
```

```python
df[postColumn].plot(kind='hist', bins=5, edgecolor='black')  # Customize bins, color, etc.
plt.title('Histogram of ' + postColumn)
plt.xlabel(postColumn)
plt.ylabel('Frequency')
plt.show()
```

```python
# Shapiro-Wilk Test
stat, p_value = shapiro(df[postColumn])

# Interpretation
def is_normal(p_value):
    alpha = 0.05
    if p_value > alpha:
        print("Sample looks normal (fail to reject H0)")
    else:
        print("Sample does not look normal (reject H0)")

print(f"Shapiro-Wilk Test Statistic for {postColumn}:", stat)
print("p-value:", p_value)
```

```python
is_normal(p_value)
```

```python
# Shapiro-Wilk Test
stat, p_value = shapiro(df[preColumn])

print(f"Shapiro-Wilk Test Statistic for {preColumn}:", stat)
print("p-value:", p_value)
```

```python
is_normal(p_value)
```
